## Load Three Models

Run each cell below to load a different model. This might take a minute -- you're downloading three AI brains!

```python
# Model 1: Binary (trained on movie reviews -- only knows POSITIVE and NEGATIVE)
model_binary = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
print("Binary model loaded!")
```

```python
# Model 2: 7 Emotions (trained on tweets -- knows anger, disgust, fear, joy, neutral, sadness, surprise)
model_emotion = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")
print("Emotion model loaded!")
```

```python
# Model 3: Zero-shot (uses whatever feeling labels YOU give it)
model_zeroshot = pipeline("zero-shot-classification", model="valhalla/distilbart-mnli-12-3")
print("Zero-shot model loaded!")
```

## Same Text, Three Different Feelings

Run the same sentence through all three models and compare.

```python
text = "I stayed up until 4am thinking about what I said to her. I keep replaying it."

# Binary: POSITIVE or NEGATIVE
r1 = model_binary(text)[0]
print(f"BINARY: {r1['label']} ({r1['score']:.0%})")

# Emotion: one of 7 emotions
r2 = model_emotion(text)[0]
print(f"EMOTION: {r2['label']} ({r2['score']:.0%})")

# Zero-shot: your custom labels
r3 = model_zeroshot(text, candidate_labels=["love", "grief", "anxiety", "hope", "regret", "confusion"])
print(f"\nYOUR LABELS:")
for label, score in zip(r3["labels"], r3["scores"]):
    print(f"  {label}: {score:.0%}")
```

### Experiment 3: Feelings that don't have one word

**Try this:** Some feelings are complex -- nostalgia, bittersweet, ambivalence. Can any of the three models see them?

```python
# Bittersweet / nostalgic text
text = "My grandmother used to make this recipe. Now I make it for my kids."

r1 = model_binary(text)[0]
print(f"BINARY: {r1['label']} ({r1['score']:.0%})")

r2 = model_emotion(text)[0]
print(f"EMOTION: {r2['label']} ({r2['score']:.0%})")

# Try labeling with complex feelings
r3 = model_zeroshot(text, candidate_labels=["nostalgia", "grief", "gratitude", "bittersweet", "love", "loss"])
print(f"\nCOMPLEX FEELINGS:")
for label, score in zip(r3["labels"], r3["scores"]):
    print(f"  {label}: {score:.0%}")

print("\n--- Can any model capture a feeling that's two things at once? ---")
```

### Experiment 4: Design your own feeling labels

**Try this:** The zero-shot model uses whatever labels you give it. What if you designed the perfect set of feelings for a specific kind of text?

```python
# Design your own labels for a specific type of text
# For example, labels for reading someone's diary:
my_labels = ["loneliness", "hope", "self-doubt", "determination", "numbness", "gratitude"]

texts = [
    "I don't know why I even bother anymore. Nothing I do seems to matter.",
    "Dear future me, I hope you figured it out.",
    "Today was actually okay. Not great, not terrible. Just okay.",
]

for text in texts:
    r = model_zeroshot(text, candidate_labels=my_labels)
    top = r["labels"][0]
    score = r["scores"][0]
    print(f"{top.upper()} ({score:.0%}): \"{text}\"")
    print()
```

## Vocabulary

| Term | Meaning |
|------|--------|
| **Training data** | The examples a model learned from |
| **Representation** | The categories or labels the training data uses |
| **Sentiment** | Positive or negative feeling |
| **Emotion categories** | The specific feelings a model was trained to recognize |
| **Model card** | Documentation about what a model does and how it was trained |
| **Fine-tuned** | A model further trained on specific data for a specific task |
| **Zero-shot** | A model that can classify text into categories it wasn't specifically trained on |
