## Load the 3 Class Models

These are the same three models from the session. Just run these cells -- nothing to change here.

```python
# Model 1: Binary (trained on movie reviews -- only knows POSITIVE and NEGATIVE)
model_binary = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
print("Binary model loaded!")
```

```python
# Model 2: 7 Emotions (trained on tweets -- knows anger, disgust, fear, joy, neutral, sadness, surprise)
model_emotion = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")
print("Emotion model loaded!")
```

```python
# Model 3: Zero-shot (uses whatever feeling labels YOU give it)
model_zeroshot = pipeline("zero-shot-classification", model="valhalla/distilbart-mnli-12-3")
print("Zero-shot model loaded!")
```

## Define Your 5 Test Inputs

Pick 5 sentences that you think will be interesting to test. Good inputs are:
- Emotionally complex (not just "I'm happy" or "I'm sad")
- A mix of obvious and tricky

Replace the example texts below with your own.

```python
# CHANGE THESE: Replace with 5 inputs you want to test
test_inputs = [
    "Replace this with your first test sentence.",
    "Replace this with your second test sentence.",
    "Replace this with your third test sentence.",
    "Replace this with your fourth test sentence.",
    "Replace this with your fifth test sentence.",
]

# Zero-shot labels -- customize these to match what you're testing
zeroshot_labels = ["love", "grief", "anxiety", "hope", "anger", "confusion"]

print(f"Ready to test {len(test_inputs)} inputs across 4 models.")
```

## Run All 4 Models on Each Input

This cell runs every model on every input and prints a comparison table. Just run it.

```python
for i, text in enumerate(test_inputs, 1):
    print(f"\n{'='*60}")
    print(f"INPUT {i}: \"{text}\"")
    print(f"{'='*60}")
    
    # Binary
    r1 = model_binary(text)[0]
    print(f"  Binary:      {r1['label']:>10}  ({r1['score']:.0%})")
    
    # Emotion
    r2 = model_emotion(text)[0]
    print(f"  Emotion:     {r2['label']:>10}  ({r2['score']:.0%})")
    
    # Zero-shot (top label only)
    r3 = model_zeroshot(text, candidate_labels=zeroshot_labels)
    print(f"  Zero-shot:   {r3['labels'][0]:>10}  ({r3['scores'][0]:.0%})")
    
    # Your model
    r4 = my_model(text)[0]
    print(f"  YOUR MODEL:  {r4['label']:>10}  ({r4['score']:.0%})")
```

## Summary Table

This cell prints a compact table so you can see all the results at once.

```python
print(f"{'Input':<6} {'Binary':<14} {'Emotion':<14} {'Zero-shot':<14} {'YOUR MODEL':<14}")
print("-" * 62)

for i, text in enumerate(test_inputs, 1):
    r1 = model_binary(text)[0]
    r2 = model_emotion(text)[0]
    r3 = model_zeroshot(text, candidate_labels=zeroshot_labels)
    r4 = my_model(text)[0]
    
    print(f"  {i:<4} {r1['label']:<14} {r2['label']:<14} {r3['labels'][0]:<14} {r4['label']:<14}")

print()
print("Which model agreed with YOUR reading most often?")
```
