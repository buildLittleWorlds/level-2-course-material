## Vocabulary

| Term | Meaning |
|------|---------|
| **Supervised learning** | Training with labeled examples where humans provide the correct answers |
| **Task** | The type of problem a model was trained to solve (classification, generation, etc.) |
| **Model Hub** | The Hugging Face library of open-source models, searchable by task |
| **Pre-trained** | A model that already learned from data — ready to use without training |
| **Fine-tuning** | Taking a pre-trained model and training it further on specific data |
