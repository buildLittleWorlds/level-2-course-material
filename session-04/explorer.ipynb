## Run the Comparison

This runs every test input through all your models and prints a comparison table.

```python
# Run all models on each input
print("=" * 80)
print("BASELINE COMPARISON")
print("=" * 80)

for i, text in enumerate(test_inputs, 1):
    print(f"\n--- Input {i}: \"{text[:60]}{'...' if len(text) > 60 else ''}\" ---")
    
    result_a = model_a(text)[0]
    result_b = model_b(text)[0]
    
    print(f"  Model A: {result_a['label']:>20s}  ({result_a['score']:.0%})")
    print(f"  Model B: {result_b['label']:>20s}  ({result_b['score']:.0%})")
    
    # Uncomment if you loaded model_c
    # result_c = model_c(text)[0]
    # print(f"  Model C: {result_c['label']:>20s}  ({result_c['score']:.0%})")
```
