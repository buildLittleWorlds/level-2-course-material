## Load a Generator

Now load a **text generation model**. If you don't have one in your Collection yet, use `distilbert/distilgpt2` (the same one from class).

```python
# CHANGE THIS if you have a different generator in your Collection
# Or keep distilgpt2 as the default

GENERATOR_NAME = "distilbert/distilgpt2"

my_generator = pipeline("text-generation", model=GENERATOR_NAME)
print(f"Generator loaded: {GENERATOR_NAME}")
print("\nBoth models ready.")
```

## Run the Comparison

Each input goes through both models. The classifier gives a label; the generator writes a continuation.

```python
print("CLASSIFICATION vs. GENERATION")
print("=" * 70)

for i, text in enumerate(test_inputs, 1):
    print(f"\n--- Input {i} ---")
    print(f'  "{text}"')
    print()
    
    # Classification
    c_result = my_classifier(text)[0]
    print(f"  CLASSIFIER: {c_result['label']} ({c_result['score']:.0%})")
    
    # Generation
    g_result = my_generator(text, max_new_tokens=50, do_sample=True, truncation=True)
    generated = g_result[0]['generated_text']
    # Show only the NEW text (after the prompt)
    new_text = generated[len(text):].strip()
    print(f"  GENERATOR:  ...{new_text[:120]}{'...' if len(new_text) > 120 else ''}")
    print()
```

## Bonus: The Variability Test

Pick your favorite input and run the generator on it 3 times. Does it produce the same output? The classifier will.

```python
# Pick one of your test inputs (change the index if you want)
favorite = test_inputs[0]

print(f'Input: "{favorite}"')
print()

print("CLASSIFIER (3 runs):")
for i in range(3):
    r = my_classifier(favorite)[0]
    print(f"  Run {i+1}: {r['label']} ({r['score']:.0%})")

print()
print("GENERATOR (3 runs):")
for i in range(3):
    r = my_generator(favorite, max_new_tokens=40, do_sample=True, truncation=True)
    new_text = r[0]['generated_text'][len(favorite):].strip()
    print(f"  Run {i+1}: ...{new_text[:100]}")
    print()

print("Classification: same every time. Generation: different every time.")
print("Next week we learn to CONTROL that variability with temperature.")
```

## The Bigger Picture

You just ran the same inputs through two fundamentally different kinds of AI.

**Classification** needs labeled data — a human decides "this is positive," "this is spam," "this is anger." That labeling is expensive, which limits how much training data you can get.

**Generation** needs just text — no labels. The model reads text and learns to predict the next word. When you don't need labels, you can train on *the entire internet*. That's what happened. And that's how we got generative AI.

The classifier you just used was trained on labeled examples — maybe thousands, maybe millions, but each one was labeled by a person. The generator was trained on raw text from the web — no labels required. Same library, same `pipeline()` function, completely different paradigm.
