The pattern: **INPUT** (text) -> **MODEL** (sentiment analyzer) -> **OUTPUT** (mood reading)

## Load the Model

This is the same model the Space uses. Run this cell and wait for it to finish.

```python
analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
print("Model loaded!")
```

## Try It Out

Run this cell to check the mood of a sentence. The model gives a label (POSITIVE or NEGATIVE) and a confidence score.

```python
result = analyzer("I can't believe how lucky I am to have friends like you.")[0]

print(f"Label: {result['label']}")
print(f"Confidence: {result['score']:.0%}")
```

## Gather Your Evidence

Each experiment below tests the model on a different kind of text. Pay attention to when you agree with the model and when you don't -- that's your evidence.

### Experiment 1: Happy vs. Sad

**Try this:** Run both cells below. Does the model get these right?

```python
# Something clearly happy
result = analyzer("Today was the best day of my life! Everything went perfectly.")[0]
print(f"{result['label']} ({result['score']:.0%})")
```

```python
# Something clearly sad
result = analyzer("I don't know why I even bother anymore. Nothing I do seems to matter.")[0]
print(f"{result['label']} ({result['score']:.0%})")
```

### Experiment 2: Ambiguous Text

**Try this:** What about text that could go either way? Does the confidence score drop?

```python
# Ambiguous -- could be happy or sad depending on how you read it
texts = [
    "Walking into school on the first day felt like stepping onto another planet.",
    "Thanks for letting me know.",
    "Dear future me, I hope you figured it out.",
    "It is what it is.",
]

for text in texts:
    result = analyzer(text)[0]
    print(f"{result['label']} ({result['score']:.0%}) -- \"{text}\"")
    print()
```

**Evidence log:** Did the confidence score change for ambiguous text? Did you agree with the model's labels? Record what you noticed -- this is the kind of evidence that matters.
