```python
# Try changing how the result is displayed
def format_for_audience(label, score):
    # CHANGE THIS to match your audience:
    if label == "POSITIVE":
        return f"Positive vibes! ({score:.0%} sure)"
    else:
        return f"Hmm, not great. ({score:.0%} sure)"

# Test it:
result = analyzer("This is the best class ever!")[0]
print(format_for_audience(result["label"], result["score"]))
```

```python
# Now try it with a few different inputs
test_inputs = [
    "The food was absolutely delicious and the service was perfect.",
    "I waited 30 minutes and the order was wrong.",
    "It was fine. Nothing special.",
]

for text in test_inputs:
    result = analyzer(text)[0]
    print(f"Input: {text}")
    print(f"Output: {format_for_audience(result['label'], result['score'])}")
    print("---")
```
