## Load Your Model

Text generation models have the most obvious parameters to sweep (temperature, top-p, max length).

`distilgpt2` is a good default -- it's small and fast. Swap it for another text-generation model from your Collection if you have one.

```python
# Swap the model name if you have a text-generation model in your Collection
generator = pipeline("text-generation", model="distilgpt2")
print("Model loaded!")
```

## Define Your Prompt

Pick one prompt and stick with it for the whole sweep. That way the only thing changing is the parameter.

```python
# Replace this with your own prompt
prompt = "The best thing about learning AI is"
```

## Sweep 1: Temperature

Temperature controls randomness. Low = predictable and safe. High = creative and wild.

Watch how the output changes as temperature goes up.

```python
temperatures = [0.3, 0.7, 1.0, 1.3, 1.7]

print("=" * 70)
print(f"TEMPERATURE SWEEP")
print(f"Prompt: \"{prompt}\"")
print("=" * 70)

for temp in temperatures:
    result = generator(
        prompt,
        max_new_tokens=50,
        temperature=temp,
        do_sample=True,
        top_p=1.0,  # keep top_p fixed
    )
    text = result[0]["generated_text"]
    print(f"\n--- Temperature = {temp} ---")
    print(text)
```

## Sweep 2: Top-p (Nucleus Sampling)

Top-p controls how many word choices the model considers. Low = only the most likely words. High = more variety.

Temperature is fixed at 1.0 so we only change one thing at a time.

```python
top_p_values = [0.5, 0.7, 0.9, 0.95, 1.0]

print("=" * 70)
print(f"TOP-P SWEEP (temperature fixed at 1.0)")
print(f"Prompt: \"{prompt}\"")
print("=" * 70)

for p in top_p_values:
    result = generator(
        prompt,
        max_new_tokens=50,
        temperature=1.0,  # keep temperature fixed
        do_sample=True,
        top_p=p,
    )
    text = result[0]["generated_text"]
    print(f"\n--- top_p = {p} ---")
    print(text)
```

## Combine Your Best Settings

Pick your favorite temperature from Sweep 1 and your favorite top-p from Sweep 2. Try them together.

```python
# Fill in your best settings
best_temperature = 0.7   # <-- change this to your favorite
best_top_p = 0.9         # <-- change this to your favorite

print(f"Combined: temperature={best_temperature}, top_p={best_top_p}")
print(f"Prompt: \"{prompt}\"")
print("-" * 70)

# Run it 3 times to see how consistent it is
for run in range(1, 4):
    result = generator(
        prompt,
        max_new_tokens=50,
        temperature=best_temperature,
        do_sample=True,
        top_p=best_top_p,
    )
    text = result[0]["generated_text"]
    print(f"\nRun {run}: {text}")
```
